- 데이터 : deliver.csv (날씨정보에 따른 배달품목별 주문량 데이터) / '내 드라이브'에 사본 저장할 것.

1. '날짜' 컬럼을 보고, '월' 컬럼과 '계절' 컬럼을 새로 추가하라.
(계절 컬럼 : 3-5월 : 봄, 6-8월 : 여름, 9-11월: 가을, 12-2월 : 겨울)

2. '미세먼지(㎍/㎥)' 컬럼에서 null값의 개수를 확인하고, null값을 전체 평균으로 대체하라.
(전체 평균은 null값이 아닌 값들의 평균으로 계산된다.)

```python
deliver = pd.read_csv('/content/gdrive/MyDrive/ybigta/deliver.csv')
deliver.head()
```

```python
deliver.info()
```

```python
import datetime as dt
deliver['날짜'] = pd.to_datetime(deliver['날짜'])
```

```python
deliver['월'] = deliver['날짜'].dt.month
deliver.head()
```

```python
def season(x) :
  if 3<=x<=5 :
    return '봄'
  elif  6<=x<=8 : 
    return '여름'
  elif 9<=x<=11 :
    return '가을'
  else :
    return '겨울'
```

```python
deliver['계절'] = deliver['월'].apply(season)
deliver.head()
```

```python
deliver[deliver['미세먼지(㎍/㎥)'].isnull()].head(20)
```

```python
deliver['미세먼지(㎍/㎥)'].fillna(deliver['미세먼지(㎍/㎥)'].mean(),inplace=True)
deliver.head(30)
```
